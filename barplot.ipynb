```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(1,11)

np.random.seed(101)

y=np.random.randint(1,100,10)

ytemp=np.random.randint(1,100,10)

age=np.random.randint(1,100,30)

bins=10

range=(0,100)

print(age)
type(age)

tick_name=['one', 'two','three','four','five','six','seven','eight','nine','ten']

plt.plot(x,y,label='Main')
plt.plot(x,ytemp,label='Temp')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()

plt.show()


plt.bar(x,y,tick_label=tick_name,color='Red',edgecolor='green')
plt.title("Freqency of year 2020")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()


plt.scatter(x,y,color='black',label='Ages')

plt.legend()
plt.show()

plt.hist(age,bins,range,histtype='bar',rwidth=0.7)
plt.title("Histogram of ages")
plt.show()


activites = ['eat','sleep','work','play']

slices = [3,7,8,6]

colors = ['r','g','m','b']


plt.pie(slices, labels=activites, colors=colors,startangle=90,shadow=True,
       explode=(0.2,0,0,0),autopct='%1.2f%%')

plt.legend()

plt.show()


xrange=np.arange(0,2*np.pi,0.1)

x1=np.sin(xrange)
y1=np.cos(xrange)

plt.plot(xrange,x1,label='Sin')
plt.plot(xrange,y1,label='Cos')

plt.legend()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
print(xrange)
print(x1)
print(y1)
```
