```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeRegressor
```

```python
data=pd.read_csv("CarPrice_Data.csv")
data
```

```python
X=data["horsepower"].values
X=X.reshape(-1,1)
```

```python
Y=data["price"].values
Y=Y.reshape(-1,1)
```

```python
plt.scatter(X,Y,color="blue")
```

```python
training_x,testing_x,training_y,testing_y= train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
reg=DecisionTreeRegressor(random_state=0)
reg.fit(training_x,training_y)
```

```python
Predict_Y=reg.predict(testing_x)
```

```python
Predict_Y
```

```python
testing_y
```

```python
X_grid = np.arange(min(testing_x), max(testing_x), 0.01) 
```

```python
X_grid = X_grid.reshape((len(X_grid), 1))
```

```python
plt.scatter(testing_x, testing_y, color = 'red') 
plt.plot(X_grid, reg.predict(X_grid), color = 'blue') 
plt.title('Horse Power to Price (Decision Tree Regression)')  
  
# specify X axis label 
plt.xlabel('Horse Power') 
  
# specify Y axis label 
plt.ylabel('Price of Car') 
  
# show the plot 
plt.show() 
```
