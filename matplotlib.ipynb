```python
import matplotlib.pyplot as pplot
import numpy as np
```

```python
x=np.arange(1,11)
y=np.random.randint(20,30,10)

print(x)
print(y)
```

```python
pplot.plot(x,y)
pplot.xlabel('X-Axis')
pplot.ylabel('Y-Axis')
pplot.title('Our Stock Price')
pplot.show()
```

```python
t = np.arange(0.0,2.0,0.01)


s = 1 + np.cos(2*np.pi*t)

pplot.plot(t,s,'--')

pplot.grid()
pplot.xlabel('Time(t)')
pplot.ylabel('Voltage(mV)')
pplot.title('Cosine wave plot(Cos(x))')

pplot.show()
```
